```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns



# Load data
df = pd.read_csv('../data/sensor_data.csv', parse_dates=['timestamp'])

# Display basic statistics
print(df.describe())

# Plot temperature over time
plt.figure(figsize=(10, 5))
plt.plot(df['timestamp'], df['temperature'])
plt.title('Temperature Over Time')
plt.xlabel('Timestamp')
plt.ylabel('Temperature')
plt.xticks(rotation=45)
plt.show()

# Additional exploratory analysis
# Plot vibration over time
plt.figure(figsize=(10, 5))
plt.plot(df['timestamp'], df['vibration'], color='orange')
plt.title('Vibration Over Time')
plt.xlabel('Timestamp')
plt.ylabel('Vibration')
plt.xticks(rotation=45)
plt.show()

# Plot pressure over time
plt.figure(figsize=(10, 5))
plt.plot(df['timestamp'], df['pressure'], color='green')
plt.title('Pressure Over Time')
plt.xlabel('Timestamp')
plt.ylabel('Pressure')
plt.xticks(rotation=45)
plt.show()

# Maintenance required distribution
plt.figure(figsize=(5, 5))
sns.countplot(x='maintenance_required', data=df)
plt.title('Maintenance Required Distribution')
plt.xlabel('Maintenance Required')
plt.ylabel('Count')
plt.show()

# Correlation heatmap
plt.figure(figsize=(10, 5))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap')
plt.show()
```
